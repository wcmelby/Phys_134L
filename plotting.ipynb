```python
temperature = np.linspace(300, 1000, 100)  # Generate 100 temperature values from 300K to 1000K
values = np.random.rand(100) * 100  # Generate 100 random values for L/L0

# Plot the data points with colors corresponding to temperature
plt.scatter(temperature, values, c=temperature, cmap='coolwarm')
plt.xlabel('Temperature (K)', fontweight='bold', fontsize=14)
plt.ylabel('L/L0', fontweight='bold', fontsize=14)
plt.legend(fontsize=12)
# plt.colorbar(label='Temperature (K)')  # Add a color bar to show the temperature scale
plt.show()
```
