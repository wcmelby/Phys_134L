```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
values = [17000, 16265.0, 17100, 17100.0]      # The four data points
errors = [0.1, 278, 260, 256]   # Corresponding error values
labels = ['Our Measurement', "O'Brien (2024)", 'Tremblay (2017)', 'Gianninas (2011)']

# Create an array for the x positions
x_positions = np.arange(len(values))  # This will be [0, 1, 2, 3]

# Create the figure and axis
fig, ax = plt.subplots()

# Plot the points with error bars
# fmt='o' indicates a circular marker
ax.errorbar(x_positions, values, yerr=errors, fmt='o', capsize=5, markersize=8, linewidth=2)

# Set the x-axis tick labels
ax.set_xticks(x_positions)
ax.set_xticklabels(labels, rotation=45, ha='right')

# Optionally add labels and a title
ax.set_ylabel('Temperature (K)')
ax.set_title('* omi02 Eri B')

# Adjust the layout so that tick labels fit well
#plt.tight_layout()
plt.grid(True)

# Show the plot
plt.savefig('omi02 Eri B')
```
